```python
# Import of libraries required for webscraping
from bs4 import BeautifulSoup
import pandas as pd
import requests

# Set URL of website in a variable
url = 'https://www.scrapethissite.com/pages/forms/'
page = requests.get(url)
print(page)
```

```python
# Use BeautifulSoup to extract the HTML Content
soup =  BeautifulSoup(page.text, 'html.parser')
print(soup)
```

```python
# Extract the table with the Hockey Scores
hockey_table = soup.find('table', class_='table')
print(hockey_table)
```

```python
# Extract the column headings
table_titles = hockey_table.find_all('th')
hockey_table_title = [title.text.strip() for title in table_titles]
print(hockey_table_title)
```

```python
# Save the column headings into a Pandas DataFrame
df = pd.DataFrame(columns=hockey_table_title)
df
```

```python
# Extract the data row by row.
# First get all rows, then loop through each row
# while stripping and saving data into the DataFrame
table_raw_data = hockey_table.find_all('tr')
for row in table_raw_data[1:]:
  raw_data = row.find_all('td')
  each_raw_data = [each_data.text.strip() for each_data in raw_data]
  print(each_raw_data)
  # saving each row data as it is generated into the pandas dataframe
  new_rows = pd.DataFrame([each_raw_data], columns=hockey_table_title)
  df = pd.concat([df, new_rows], ignore_index=True)
```

```python
# inspect the resulting dataframe
df
```

```python
# Save to a .csv file in the same folder
df.to_csv(r'./Hockey.csv')
```
